```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
all_data = pd.DataFrame()
```

```python
for page_number in range(1, 2001):  
    url = f'https://erp.pgcb.gov.bd/w/generations/view_generations?page={page_number}'
    
    response = requests.get(url,verify=False)
    
    # Checking if the page is accessible
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        table = soup.find('table')

        # Checking if the table is found on the page
        if table:
            page_data = pd.read_html(str(table))[0]
            all_data = all_data.append(page_data, ignore_index=True)
```

```python
all_data.to_excel('combined_data.xlsx', index=False)
```
